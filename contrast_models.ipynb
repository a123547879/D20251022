```python
import torch
import torchvision
from torch import nn, optim
from torch.utils.data import DataLoader
from matplotlib import pyplot as plt
from torchvision import transforms, datasets
from cnn_model2 import PrecisionBalancedCNN
from GhostNet import GhostNet
```

```python
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)  # 如果使用多GPU
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed = 42
# 在训练代码开头调用
set_seed(seed) # 42, 2025, 1024, 512, 100
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307, ), (0.3081, )),
    transforms.Grayscale(1)
])
```

```python
test_db = datasets.ImageFolder(root= 'C:/Users/Administrator/Desktop/datas/test_data_black', transform= transform)
test_loader = DataLoader(test_db, batch_size= 10, shuffle= True)
len(test_loader)
```

```python
model_basic = PrecisionBalancedCNN('basic', in_channels= 1)
model_basic.load_state_dict(torch.load(f'models/2/model_basic_2_{seed}.pth'))
model_basic.to(device)
model_basic.eval()
print('basic load')
```

```python
model_ghost = PrecisionBalancedCNN('ghost', in_channels= 1)
model_ghost.load_state_dict(torch.load(f'models/2/model_ghost_2_{seed}.pth'))
model_ghost.to(device)
model_ghost.eval()
print('ghost load')
```

```python
model_inverted_residual = PrecisionBalancedCNN('inverted_residual', in_channels= 1,)
model_inverted_residual.load_state_dict(torch.load(f'models/2/model_inverted_residual_2_{seed}.pth'))
model_inverted_residual.to(device)
model_inverted_residual.eval()
print('inverted_residual load')
```

```python
model_residual = PrecisionBalancedCNN('residual', in_channels= 1)
model_residual.load_state_dict(torch.load(f'models/2/model_residual_2_{seed}.pth'))
model_residual.to(device)
model_residual.eval()
print('residual load')
```

```python
model_shuffle = PrecisionBalancedCNN('shuffle', in_channels= 1)
model_shuffle.load_state_dict(torch.load(f'models/2/model_shuffle_2_{seed}.pth'))
model_shuffle.to(device)
model_shuffle.eval()
print('shuffle load')
```

```python
model_selayer = PrecisionBalancedCNN('selayer', 1)
model_selayer.load_state_dict(torch.load(f'models/2/model_selayer_2_{seed}.pth'))
model_selayer.to(device)
model_selayer.eval()
print('selayer load')
```

```python
total = 0
idxs = []
correct1, correct2, correct3, correct4, correct5, correct6 = 0, 0, 0, 0, 0, 0
correct1s, correct2s, correct3s, correct4s, correct5s, correct6s = [], [], [], [], [], []
with torch.no_grad():
    for idx, (x, y) in enumerate(test_loader):
        x, y = x.to(device), y.to(device)
        out1 = torch.argmax(model_basic(x), dim= -1)
        out2 = torch.argmax(model_ghost(x), dim= -1)
        out3 = torch.argmax(model_inverted_residual(x), dim= -1)
        out4 = torch.argmax(model_residual(x), dim= -1)
        out5 = torch.argmax(model_shuffle(x), dim= -1)
        out6 = torch.argmax(model_selayer(x), dim= -1)

        total += y.size(0)
        correct1 += (out1 == y).sum().item()
        correct2 += (out2 == y).sum().item()
        correct3 += (out3 == y).sum().item()
        correct4 += (out4 == y).sum().item()
        correct5 += (out5 == y).sum().item()
        correct6 += (out6 == y).sum().item()

        if idx % 50 == 0 or (idx + 1) == len(test_loader):
            acc1 = correct1 / total
            acc2 = correct2 / total
            acc3 = correct3 / total
            acc4 = correct4 / total
            acc5 = correct5 / total
            acc6 = correct6 / total

            idxs.append(idx)
            correct1s.append(acc1), correct2s.append(acc2), correct3s.append(acc3)
            correct4s.append(acc4), correct5s.append(acc5), correct6s.append(acc6)

            print(f'idx: {idx}, basic:{acc1:.4f}, ghost:{acc2:.4f}, inverted:{acc3:.4f}, residual:{acc4:.4f}, shuffle:{acc5:.4f}, selayer:{acc6:.4f}')
```
